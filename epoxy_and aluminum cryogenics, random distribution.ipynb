```python
# taking random matrices distribution of 10 percent Aluminium and 90 percent epoxy and finding the thermal conductivity in composite
#materials, assuming particles to be no overlap
lst=[]
lst1=[]
for i in range(100):
    a= np.random.choice([0, 1], size=(50,50), p=[9./10, 1./10])
    b=a.sum(axis=0)
    c=b.argmax()
    b[c]
    r=50
    no_Al=b[c]
    k = (1/r)*((no_Al/237)+((r - no_Al)/1.052))
    ki = 1/k
    lst.append(ki)
    c=b.argmin()
    b[c]
    r=50
    no_Al=b[c]
    k = (1/r)*((no_Al/237)+((r - no_Al)/1.052))
    ki = 1/k
    lst1.append(ki)
lst7=[]
for i in range(100):
  lst7.append(np.mean(lst[:i+1]))
lst2=[i for i in range(100)]
lst3=[i for i in range(100)]

plt.plot(lst3, lst1)
plt.plot(lst2, lst)
#Graph showing maximum and mimimum thermal conducitivity that can be reached in 100 iterations



```
